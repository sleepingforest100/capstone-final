# First Example

```java
using System;
using System.Windows.Forms;
using System.Xml;

public class CarDisplayForm : Form
{
    // [Hungarian] Notation for UI controls:
    private TextBox txtName;
    private Label lblCarColor;

    // [Camel] Notation for local variables:
    private string myCar;
    private string backColor;

    // [Acronyms] identifier using Pascal Notation for a class instance:
    private XmlWriter XmlWriterInstance;
    public CarDisplayForm()
    {
        // Initialize variables ([Camel] notation)
        myCar = "Toyota";
        backColor = "Blue";

        // Initialize UI controls ([Hungarian] notation)
        txtName = new TextBox() { Left = 20, Top = 20, Width = 200 };
        lblCarColor = new Label() { Left = 20, Top = 60, Width = 200 };

        // Set up the form
        this.Text = "Car Display";
        this.Controls.Add(txtName);
        this.Controls.Add(lblCarColor);

        // Use [Pascal] Notation for method call:
        txtName.Text = myCar;
        lblCarColor.Text = ShowCarColor();
    }

    // [Pascal] Notation for methods:
    public string ShowCarColor()
    {
        return backColor;
    }

    // [Pascal] Notation for properties:
    public string CarName { get; set; }
    public static void Main()
    {
        Application.EnableVisualStyles();
        Application.Run(new CarDisplayForm());
    }
}
```
